```python
import requests
from datetime import datetime
import pandas as pd

top_3 = ["bitcoin","ethereum","tether"]
all_data = []

for coin_id in top_3:
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=inr&days=365"

    data = requests.get(url)
    if data.status_code == 200:

        hist_data = data.json()

        if "prices" in hist_data:
           for price in hist_data["prices"]:
            timestamp = price[0]
            price_inr = price[1]

            date = datetime.fromtimestamp(timestamp / 1000).strftime("%Y-%m-%d")

            all_data.append({
                "coin_id": coin_id,
                "date": date,
                "price_inr": price_inr
            })
```

```python
price_df = pd.DataFrame(all_data)
price_df
```

```python
import pandas as pd
price_df = pd.DataFrame(all_data)
price_df.to_csv("top3_daily_price_1year_inr.csv",index=False)
```

# 1.Cryptocurrencies Table created

```python
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE Cryptocurrencies (
       id VARCHAR(50) PRIMARY KEY, ---- Unique ID from CoinGecko (e.g., “bitcoin”)
       symbol VARCHAR(50), ---Short symbol (e.g., BTC, ETH)
       name  VARCHAR(100), ---Full name of the cryptocurrency
       current_price DECIMAL(18, 6), --Current trading price INR
       market_cap BIGINT, ----Market capitalization (INR)
       market_cap_rank INT, ---Rank by market capitalization
       total_volume BIGINT, ---Trading volume (24h, INR)
       circulating_supply DECIMAL(20, 6), ---Coins in circulation
       total_supply DECIMAL(20, 6), ---Total supply (if available)
       ath DECIMAL(18, 6), ----All-time high price (INR)
       atl DECIMAL(18, 6),---All-time low price (INR)
       date DATE ----Last updated
    )
''')
```

# Insert into Table

```python
import pandas as pd
pd.read_sql("""SELECT name, market_cap
    FROM Cryptocurrencies
    ORDER BY market_cap DESC
    LIMIT 3;""", conn)
```

# 2.Get the average oil price per year.

```python
pd.read_sql("""SELECT
   id,symbol,name,circulating_supply,total_supply,
   ROUND((circulating_supply * 100.0) / total_supply, 2) AS supply_percentage
   FROM Cryptocurrencies
   WHERE total_supply IS NOT NULL
   AND total_supply > 0
   AND circulating_supply >= 0.9 * total_supply
   ORDER BY supply_percentage DESC;
   """,conn)
```

# 3.Show oil prices during COVID crash (March–April 2020).

```python
pd.read_sql("""SELECT
    id,symbol,name,current_price,ath,
    ROUND((current_price / ath) * 100, 2) AS percent_of_ath
    FROM Cryptocurrencies
    WHERE ath IS NOT NULL
    AND ath > 0
    AND current_price >= 0.9 * ath
    ORDER BY percent_of_ath DESC;
    """, conn)
```

# 15.Calculate the volatility of oil prices (max-min difference per year).

```python
pd.read_sql("""SELECT
  strftime('%Y', date) AS year,
  MAX(price_inr) AS max_price,
  MIN(price_inr) AS min_price,
  ROUND(MAX(price_inr) - MIN(price_inr), 2) AS volatility
  FROM oil_prices
  GROUP BY year
  ORDER BY year;
  """,conn)
```

```python
%%writefile app.py
import streamlit as st
import sqlite3
import pandas as pd
import datetime as dt
conn = sqlite3.connect("Market.db")
st.sidebar.title("Navigation")

menu = st.sidebar.selectbox(
    "Select a page:",
    ["Home", "Cross Market Overview", "SQL Query Runner", "Top 3 Crypto Analysis"]
)

# -------------------------
# HOME PAGE
# -------------------------
if menu == "Home":
    st.title("📊 Cross Market Analysis")
    st.write("Welcome to the Cross Market Analysis Dashboard")

# -------------------------
# CROSS MARKET OVERVIEW
# -------------------------
elif menu == "Cross Market Overview":
    st.title("📈 Cross Market Overview")


    st.subheader("Select Date Range")
    start_date = st.date_input("Start Date", value=pd.to_datetime("2020-01-01"))
    end_date = st.date_input("End Date", value=pd.to_datetime("today"))

    btc_q = f"""
    SELECT date, current_price AS bitcoin_price
    FROM Cryptocurrencies
    WHERE date BETWEEN '{start_date.strftime('%Y-%m-%d')}' AND '{end_date.strftime('%Y-%m-%d')}'
    """

    oil_q = f"""
    SELECT date, price_inr AS oil_price
    FROM oil_prices
    WHERE date BETWEEN '{start_date.strftime('%Y-%m-%d')}' AND '{end_date.strftime('%Y-%m-%d')}'
    """

    sp_q = f"""
    SELECT date, close AS sp500_close
    FROM stock_prices
    WHERE ticker = '^GSPC'
      AND date BETWEEN '{start_date.strftime('%Y-%m-%d')}' AND '{end_date.strftime('%Y-%m-%d')}'
    """

    nifty_q = f"""
    SELECT date, close AS nifty_close
    FROM stock_prices
    WHERE ticker = '^NSEI'
      AND date BETWEEN '{start_date.strftime('%Y-%m-%d')}' AND '{end_date.strftime('%Y-%m-%d')}'
    """

    btc_df = pd.read_sql(btc_q, conn)
    oil_df = pd.read_sql(oil_q, conn)
    sp_df = pd.read_sql(sp_q, conn)
    nifty_df = pd.read_sql(nifty_q, conn)

    if not btc_df.empty and not oil_df.empty and not sp_df.empty and not nifty_df.empty:
        col1, col2, col3, col4 = st.columns(4)
        col1.metric("Bitcoin Avg", round(btc_df["bitcoin_price"].mean(), 2))
        col2.metric("Oil Avg", round(oil_df["oil_price"].mean(), 2))
        col3.metric("S&P 500 Avg", round(sp_df["sp500_close"].mean(), 2))
        col4.metric("NIFTY Avg", round(nifty_df["nifty_close"].mean(), 2))

        # -------------------------
        # 📅 Daily Market Snapshot (Separate Date Ranges)
        # -------------------------
        st.subheader("📅 Daily Market Snapshot (Separate Date Ranges)")

        # --- Bitcoin ---
        st.markdown("### 🪙 Bitcoin Date Range")
        btc_start = st.date_input("Bitcoin Start Date", dt.date(2026, 3,4), key="btc_start")
        btc_end   = st.date_input("Bitcoin End Date", dt.date(2026, 3, 4), key="btc_end")

        btc_snap = pd.read_sql(f"""
        SELECT date, current_price AS bitcoin_price
        FROM Cryptocurrencies
        WHERE id = 'bitcoin'
          AND date BETWEEN '{btc_start}' AND '{btc_end}'
        ORDER BY date;
        """, conn)

        st.dataframe(btc_snap, use_container_width=True)

        # --- Oil ---
        st.markdown("### 🛢️ Oil Date Range")
        oil_start = st.date_input("Oil Start Date", dt.date(2020, 1, 2), key="oil_start")
        oil_end   = st.date_input("Oil End Date", dt.date(2026, 1, 26), key="oil_end")

        oil_snap = pd.read_sql(f"""
        SELECT date, price_inr AS oil_price
        FROM oil_prices
        WHERE date BETWEEN '{oil_start}' AND '{oil_end}'
        ORDER BY date;
        """, conn)

        st.dataframe(oil_snap, use_container_width=True)

        # --- NIFTY ---
        st.markdown("### 📈 NIFTY Date Range")
        nifty_start = st.date_input("NIFTY Start Date", dt.date(2020, 1, 1), key="nifty_start")
        nifty_end   = st.date_input("NIFTY End Date", dt.date(2026, 1, 9), key="nifty_end")

        nifty_snap = pd.read_sql(f"""
        SELECT date, open, high, low, close, volume
        FROM stock_prices
        WHERE ticker = '^NSEI'
          AND date BETWEEN '{nifty_start}' AND '{nifty_end}'
        ORDER BY date;
        """, conn)

        st.dataframe(nifty_snap, use_container_width=True)

    else:
        st.warning("No data available for selected date range.")
# -------------------------
# SQL QUERY RUNNER
# -------------------------
elif menu == "SQL Query Runner":
    st.title("🔎 SQL Query Runner")

    conn = sqlite3.connect("Market.db")

    queries = {
       "1. Find the top 3 cryptocurrencies by market cap":
        "SELECT name, market_cap FROM cryptocurrencies ORDER BY market_cap DESC LIMIT 3;",

    "2. List all coins where circulating supply exceeds 90% of total supply":
        "SELECT name, circulating_supply, total_supply FROM cryptocurrencies WHERE circulating_supply >= 0.9 * total_supply;",

    "3. Get coins that are within 10% of their all-time-high (ATH)":
        "SELECT name, current_price, ath FROM cryptocurrencies WHERE current_price >= 0.9 * ath;",

    "4. Find the average market cap rank of coins with volume above 1B":
        "SELECT AVG(market_cap_rank) AS avg_rank FROM cryptocurrencies WHERE total_volume > 1000000000;",

    "5. Get the most recently updated coin":
       "SELECT id,symbol,name,date FROM Cryptocurrencies ORDER BY date DESC LIMIT 1; ",

    "6.Find the highest daily price of Bitcoin in the last 365 days.":
    "SELECT coin_id, MAX(price_inr) AS highest_price_last_365_days FROM top3_daily_price_1year_inr WHERE coin_id = 'bitcoin' AND date >= DATE('now', '-365 day');",

    "7.Calculate the average daily price of Ethereum in the past 1 year.":
    "Select  avg(price_inr)as avg_price_last_365_days from top3_daily_price_1year_inr where coin_id = 'ethereum' and date >= DATE('now', '-366 day');",

    "8.Show the daily price trend of Bitcoin in January 2025.":
    "Select date,coin_id,price_inr  from top3_daily_price_1year_inr  where coin_id = 'bitcoin' and date between '2025-01-01' and '2025-12-31'   order by date;",

    "9.Find the coin with the highest average price over 1 year.":
     "Select coin_id, avg(price_inr)as avg_price from top3_daily_price_1year_inr group by coin_id order by avg_price desc limit 1;",

    "10.Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025.":
    "SELECT ((MAX(price_inr) - MIN(price_inr)) / MIN(price_inr)) * 100 AS pct_change FROM top3_daily_price_1year_inr WHERE coin_id='bitcoin'AND date BETWEEN '2025-01-27' AND '2026-01-26';",

    "11.Find the highest oil price in the last 5 years.":
    "Select max(price_inr) as highest_price from oil_prices where date >= DATE('now', '-5 years');",
    "12.Get the average oil price per year.":
    " SELECT strftime('%Y', date) AS year, ROUND(AVG(price_inr), 2) AS avg_price FROM oil_prices GROUP BY year ORDER BY year;",

   " 13.Show oil prices during COVID crash (March–April 2020).":
    "SELECT date, price_inr FROM oil_prices WHERE date BETWEEN '2020-03-01' AND '2020-04-30' ORDER BY date;",

    "14.Find the lowest price of oil in the last 10 years.":
     "Select min(price_inr) as lowest_price from oil_prices where date >= DATE('now', '-10 years');",

    "15.Calculate the volatility of oil prices (max-min difference per year).":
    "SELECT strftime('%Y', date) AS year, MAX(price_inr) AS max_price, MIN(price_inr) AS min_price, ROUND(MAX(price_inr) - MIN(price_inr), 2) AS volatility FROM oil_prices GROUP BY year  ORDER BY year;",

    "16.Get all stock prices for a given ticker":
    "SELECT * FROM stock_prices GROUP BY ticker ORDER BY date;",

    "17.Find the highest closing price for NASDAQ (^IXIC)":
    "SELECT MAX(close) AS highest_close FROM stock_prices WHERE ticker = '^IXIC';",

    "18.List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)":
    "SELECT date, (high - low) AS price_diff FROM stock_prices WHERE ticker = '^GSPC' ORDER BY price_diff;",

    "19.Get monthly average closing price for each ticker":
    "SELECT ticker, strftime('%Y-%m', date) AS month, AVG(close) AS avg_close FROM stock_prices GROUP BY ticker, month ORDER BY ticker, month;",

    "20.Get average trading volume of NSEI in 2024":
    "SELECT AVG(volume) AS avg_volume_2024 FROM stock_prices WHERE ticker = '^NSEI' AND date BETWEEN '2024-01-01' AND '2024-12-31';",

    "21. Find the Average Bitcoin vs stock volume average price.":
    "Select  coin_id, avg(price_inr)as avg_price, avg(s.volume) AS avg_stock_volume from top3_daily_price_1year_inr t join stock_prices s on t.date = s.date where coin_id = 'bitcoin';",

    "22.Check if oil prices with S&P 500 high (correlation idea). 2025.":
    "SELECT avg(o.price_inr) as avg_oil, avg(s.high) AS avg_stock_high FROM top3_daily_price_1year_inr t JOIN oil_prices o ON t.date = o.date JOIN stock_prices s ON t.date = s.date WHERE t.date BETWEEN '2025-01-01' AND '2025-12-31';",

    "23.Find the average price and avgerage low stock price for 2025-2026.":
    "SELECT avg(price_inr) as avg_price, avg(s.low) AS avg_stock_low FROM top3_daily_price_1year_inr t JOIN stock_prices s ON t.date = s.date WHERE t.date BETWEEN '2025-01-01' AND '2026-12-31';",

    "24.compare the oil price and the average volume stock price between 2020 and 2025.":
    "SELECT o.date AS oil_date, avg(o.price_inr) as avg_price, avg(s.volume) AS avg_stock_volume FROM oil_prices o JOIN stock_prices s ON o.date = s.date WHERE s.date BETWEEN '2020-01-01' AND '2025-12-31';",

    "25.Retrieve top3 historical prices and stock ticker, volume, and date for stock data between 2024 and 2025.":
    "SELECT avg (price_inr) as price_inr, s.ticker AS stock_ticker, s.volume AS stock_volume FROM top3_daily_price_1year_inr t JOIN stock_prices s ON t.date = s.date WHERE t.date BETWEEN '2024-01-01' AND '2025-12-31' ORDER BY t.date;",

    "26.Find the average close and oil price for the stock ticker 'NSEI' between 2020 and 2025.":
    "SELECT AVG(s.close) AS avg_stock_close, AVG(price_inr) AS avg_price FROM oil_prices o JOIN stock_prices s ON o.date = s.date WHERE s.ticker = '^NSEI'   AND o.date BETWEEN '2020-01-01' AND '2025-12-31';",

    "27.Compare the monthly average close prices of stock prices with oil prices.":
    "SELECT oil_prices.date,  ticker, strftime('%Y-%m', stock_prices.date) AS month, AVG(close) AS avg_close from oil_prices join stock_prices on oil_prices.date=stock_prices.date GROUP BY ticker, month ORDER BY ticker, month;",

    "28.Find the top 5 stock tickers with the largest price difference between high and low on the same date":
    "SELECT ticker,(high - low) AS price_diff FROM top3_daily_price_1year_inr t JOIN stock_prices s ON t.date = s.date WHERE ticker = '^GSPC' ORDER BY price_diff DESC Limit 5;",

    "29.Find the top 2 tickers and close in stock prices.":
    "SELECT   s.ticker AS stock_ticker,  s.close AS stock_close FROM stock_prices s  WHERE s.ticker IN ('^NSEI', '^GSPC')  ORDER BY  s.ticker limit 10;",

    "30.Find the top 5 stock tickers with the highest price difference (calculated as high - low) for the ticker 'GSPC' (S&P 500) from the stock prices table.":
    "SELECT  ticker, (high - low) AS price_diff FROM stock_prices s WHERE ticker = '^GSPC' ORDER BY price_diff DESC LIMIT 5;"
    }

    selected_question = st.selectbox("Select a Question", list(queries.keys()))
    st.text_area("SQL Query", queries[selected_question], height=80)

    if st.button("Run Query"):
        try:
            df = pd.read_sql(queries[selected_question], conn)
            st.success("Query executed successfully!")
            st.dataframe(df)
        except Exception as e:
            st.error(f"Error: {e}")

# -------------------------
# TOP 3 CRYPTO ANALYSIS
# -------------------------
elif menu == "Top 3 Crypto Analysis":
    st.title("🪙 Top 3 Crypto Analysis")

    conn = sqlite3.connect("Market.db")

    coins = {
        "Bitcoin (BTC)": "bitcoin",
        "Ethereum (ETH)": "ethereum",
        "Tether (USDT)": "tether"
    }

    selected_coin_name = st.selectbox("Select a Coin", list(coins.keys()))
    selected_coin = coins[selected_coin_name]

    col1, col2 = st.columns(2)
    with col1:
        start_date = st.date_input("Start date", dt.date(2025, 3, 5))
    with col2:
        end_date = st.date_input("End date", dt.date(2026, 3, 4))

    if st.button("Show Price"):
        query = """
        SELECT DISTINCT date, price_inr
        FROM top3_daily_price_1year_inr
        WHERE coin_id = ?
          AND date BETWEEN ? AND ?
        ORDER BY date ASC
        """

        df = pd.read_sql(query, conn, params=(selected_coin, str(start_date), str(end_date)))

        if df.empty:
            st.warning("No data found for this period.")
        else:
            st.success(f"Daily price for {selected_coin_name}")
            st.dataframe(df)
```
