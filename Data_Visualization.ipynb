# Adult Literacy Rate for India

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,5))
df3_india = df3[df3['Country_X'] == 'India']
sns.histplot(df3_india["adult_literacy_rate_population"], bins=30, kde=True)
plt.title("Distribution of Adult Literacy Rate for India")
plt.xlabel("Adult Literacy Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

# Literacy trend in India over the years by gender

```python
import matplotlib.pyplot as plt
df2 = df2[df2["Country_Z"] == "Sri Lanka"]
plt.figure(figsize=(4,3))
plt.plot(df2["year"], df2["illiteracy_rate"], marker="o", label="Illiteracy Rate")
plt.plot(df2["year"], df2["literacy_rate"], marker="*", label="Literacy Rate")
plt.title("Literacy & Illiteracy Rates ")
plt.xlabel("Year")
plt.ylabel("Literacy Rate")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 countries with most records
top10_countries = df1['Country_A'].value_counts().nlargest(10).index

# Filter dataframe
df_top10 = df1[df1['Country_A'].isin(top10_countries)]

# Plot
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df_top10,
    x='Country_A',
    y='gdp_per_capita',
)

plt.xticks(rotation=45)
plt.title("GDP per capita Rate Distribution for Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("GDP")

plt.show()
```
