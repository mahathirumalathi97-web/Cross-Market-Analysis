# Table 1 Creation of Literacy Rates

```python
cursor=conn.cursor()
cursor.execute('''
    Create table Literacy_rates(
     Country_X varchar(50),
     code_x varchar(50),
     year int,
     adult_literacy_rate_population float,
     Country_Y varchar(50),
     code_y varchar(50),
     male float,
     female float,
     owid_region_y varchar(50)

    )
''')
```

# Insert Values

```python
insert_query = """INSERT into Literacy_rates(Country_X,code_x, year,adult_literacy_rate_population,Country_Y,code_y, male,female,owid_region_y)
 VALUES (?,?,?,?,?,?,?,?,?)"""

for i, row in df_Literacy.iterrows():
    values = (row['Country_X'],row['code_x'],row['year'],row['adult_literacy_rate_population'],row['Country_Y'],row['code_y'],
              row['male'],row['female'],row['owid_region_y'])
    cursor.execute(insert_query,values)
conn.commit()
```

# SQL QUERIES

 # literacy_rates
 1. Get top 5 countries with highest adult literacy in 2020.

```python
import pandas as pd
pd.read_sql("""SELECT country_x, adult_literacy_rate_population
    FROM literacy_rates
    WHERE year = 2020
    ORDER BY  adult_literacy_rate_population DESC
    LIMIT 5;""", conn)
```

 2. Find countries where female youth literacy < 80%

```python
pd.read_sql("""SELECT country_Y, female
    FROM literacy_rates
    WHERE female < 80
    ORDER BY  female DESC
    ;""", conn)
```

3.Average adult literacy per continent (owid region).

```python
pd.read_sql("""SELECT owid_region_y, AVG(adult_literacy_rate_population) AS avg_adult_literacy
    FROM Literacy_rates
    GROUP BY owid_region_y;""", conn)
```

# Join Queries

   10. List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).


```python
pd.read_sql("""
SELECT
g.Country_A as country, g.year,g.gdp_per_capita,g.youth_adult_yr_edu
FROM Gdp_Schooling g
where g.year = 2020
and g.youth_adult_yr_edu < 6
order by g.gdp_per_capita desc
limit 10;
""",conn)
```

  11. Show countries where the illiterate population is high despite having more than 10 average years of schooling.

```python
pd.read_sql("""
SELECT
G.Country_A as country,
g.year,g.youth_adult_yr_edu,i.illiteracy_rate
FROM Gdp_Schooling g
join illiteracy_population i
on g.code_x = i.code
and g.year = i.year
where g.youth_adult_yr_edu > 10
and i.illiteracy_rate > 20
order by i.illiteracy_rate desc;
""",conn)
```

   12. Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)


```python
pd.read_sql("""
SELECT
g.year,g.gdp_per_capita,g.literacy_rate
FROM Gdp_Schooling g
where g.Country_A = 'India'
and g.year >= 2000 and g.year <= 2020
order by g.year;
""",conn)
```

 13. Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.

```python
pd.read_sql("""
SELECT
    g.Country_A AS country,
    g.gdp_per_capita,
    l.male AS male_literacy_rate,
    l.female AS female_literacy_rate,
    (l.male - l.female) AS literacy_gap
FROM Gdp_Schooling g
JOIN Literacy_rates l
    ON g.code_x = l.code_x
    AND g.year = l.year
WHERE g.year = 2020
  AND g.gdp_per_capita > 30000
ORDER BY literacy_gap DESC;
""",conn)
```
